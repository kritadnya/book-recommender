```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar= False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")


plt.show()
```

```python
import numpy as np
books["missing_description"] = np.where(books["description"].isna(),1,0)
# if description is missing then put a 1 in the missing_description column else put a 0
# we are then going to convert the year of the book to the age of the book
books["age"] = 2023 - books["published_year"]
# we basically added these two columns based on the information from present columns 


books.columns


```

```python
columns_of_interest = ["num_pages", "age", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")
# pearson correlations are more appropriate when you are dealing with continous variables
# as you can see the "missing_description field" is binary thus we used spearman correlation
```

```python
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label":"Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
# in this heatmap we are trying to see if missing_description is related to any of the other factors 
# understand heatmaps and how to interpret them
```

```python
# find how many books dont have a description, number of pages, average rating, published year
books[(books["description"].isna()) | (books["num_pages"].isna()) | (books["average_rating"].isna()) | (books["published_year"].isna())  ]
```

```python
# there are 303 books who have the above info missing, which is under 5% thus we can eliminate them 
```

```python
book_missing = books[~(books["description"].isna()) & ~(books["num_pages"].isna()) & ~(books["average_rating"].isna()) & ~(books["published_year"].isna())  ]

# you are saying return me a data farme in which the book must not have any of the 4 fieldds missing 
# only those books that have all 4 fields present will be selected 
# using NOT will ensure this new dataframe will exclude the books that have not available marked 
```

```python
book_missing
# this df will have only those books that dont have missing values in those fields
```

```python
# the "categories" column is pretty inconsistent and has fictional characters in it too which might not be super relevant for category
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
# if you see this "categories" table the distribution is very uneven some categories are more some are less
```

```python
book_missing.loc[:,"words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(1,4),"description"]
# we are trying to display the books with very small descriptions ie that have 1-4 words
# goal is to find at what words is the description actually helpful and descriptive enough
```

```python
#  remove all that have less than 25 words and assign those that have greater than 25 words to the data frame book_missing_25_words
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
# if subtitle is missing use the title
# if subtitle is present then aggregate the title and subtitle ; convert evrything to a string bcz there might be missing values and missing values are considered as float in pandas
book_missing_25_words.loc[:,"title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
    book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join, axis=1)
))
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```
